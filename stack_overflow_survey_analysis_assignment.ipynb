```python
#Which  is most used  programming langauge (use LanguageWorkedWith column)
pd.DataFrame(df['LanguageWorkedWith'].value_counts()).head(1)
```

```python
#Display how many work remotely on pie chart  [use WorkRemote column]
arr = pd.Series(df['WorkRemote'].value_counts())
index = np.array(arr.keys())
#index
x = arr.values
```

```python
plt.pie(x, labels=index)
plt.title('Developers Working Remotely')
plt.legend(loc='center left', bbox_to_anchor=(1.5, 0.5))
plt.axis('equal')
```

```python
#Display how many male and female programmers on pie chart  [use Gender column]
gender = pd.Series(df['Gender'].value_counts())
keys = np.array(gender.keys())
val = gender.values
```

```python
exclude = [0.05 for i in range(7)]
plt.pie(val, labels=keys, explode=exclude)
plt.title('Gender Pie Chart')
plt.legend(loc='center left', bbox_to_anchor=(1.5, 0.5))
plt.axis('equal')
```
